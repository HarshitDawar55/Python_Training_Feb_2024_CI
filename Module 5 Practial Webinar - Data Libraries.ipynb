### Task 3 - Combining 2 dataframes and working on them

```python
red_wine_df["wine_type"] = "red"
white_wine_df["wine_type"] = "white"
```

```python
red_wine_df
```

```python
final_df = pd.concat([red_wine_df, white_wine_df], axis = 0).reset_index(drop=True)
final_df
```

### Task 4 - Creating new features by combining existing ones

```python
final_df["sulphur_dioxides"] = final_df["free sulfur dioxide"] + final_df["total sulfur dioxide"]
```

```python
final_df
```

```python
final_df["acids"] = final_df["fixed acidity"] + final_df["volatile acidity"] + final_df["citric acid"]
```

```python
final_df
```

```python
final_df = final_df.drop(["free sulfur dioxide", "total sulfur dioxide", "fixed acidity", "volatile acidity", "citric acid"], axis = 1)
final_df
```
